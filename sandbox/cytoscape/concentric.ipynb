```python
import json
with open("data/concentricData.json") as fi:
    json_file = json.load(fi)
```

```python
cytoscapeobj = ipycytoscape.CytoscapeWidget()
cytoscapeobj.graph.add_graph_from_json(json_file)
```

```python
cytoscapeobj.set_layout(name="concentric")
cytoscapeobj.set_style(
    [
        {
            "selector": "node",
            "style": {"height": 20, "width": 20, "background-color": "#30c9bc"},
        },
        {
            "selector": "edge",
            "style": {
                "curve-style": "haystack",
                "haystack-radius": 0,
                "width": 5,
                "opacity": 0.5,
                "line-color": "#a8eae5",
            },
        },
    ]
)
```
